```python
import re
from pathlib import PureWindowsPath


def parse_filename(path) -> pd.Series:
    p = PureWindowsPath(str(path))
    base = p.name

    # extension
    ext = base.split(".")[-1].lower() if "." in base else pd.NA

    # stem without extension
    stem = re.sub(r"\.[^.]+$", "", base).strip()

    yyyymm = pd.NA
    rest = stem

    # Pattern 1: leading YYYYMM (e.g., 202203 ...)
    m1 = re.match(r"^\s*(\d{6})\s+(.*)$", stem)
    if m1:
        yyyymm = m1.group(1)
        rest = m1.group(2).strip()
    else:
        # Pattern 2: leading YYYY MM (e.g., 2023 10_...)
        m2 = re.match(r"^\s*(\d{4})\s+(\d{1,2})[ _\-]+(.*)$", stem)
        if m2:
            year = m2.group(1)
            month = int(m2.group(2))
            yyyymm = f"{year}{month:02d}"
            rest = m2.group(3).strip()
        else:
            # Pattern 3: leading short numeric ID (e.g., 2310 JD ...)
            m3 = re.match(r"^\s*(\d{2,6})\s+(.*)$", stem)
            if m3:
                rest = m3.group(2).strip()

    title_raw = rest

    # ---------------- Title cleaning ----------------
    t = title_raw

    # normalize separators / mojibake
    t = t.replace("‚Äì", "-").replace("–", "-")
    t = re.sub(r"[_]+", " ", t)

    # remove posting phrases
    t = re.sub(r"(?i)\binternal job posting\b", " ", t)
    t = re.sub(r"(?i)\bjob posting\b", " ", t)
    t = re.sub(r"(?i)\bposting\b", " ", t)

    # NEW: remove AD / JD / job advertisement
    t = re.sub(r"(?i)\bjob\s+advertisement\b", " ", t)
    t = re.sub(r"(?i)\bAD\b", " ", t)   # AD as a whole word
    t = re.sub(r"(?i)\bJD\b", " ", t)   # JD as a whole word

    # remove temp/contract/term language
    t = re.sub(r"(?i)\btemp\b|\btemporary\b", " ", t)
    t = re.sub(r"(?i)\bcontract\b|\bterm\b|\bsecondment\b", " ", t)

    # remove durations like "4-month", "~18-month", "12-18 month", "(14 month assignment)"
    t = re.sub(r"(?i)\(?\s*~?\s*\d+\s*(?:-\s*\d+)?\s*[- ]?\s*month[s]?\s*\)?", " ", t)
    t = re.sub(r"(?i)\(?\s*\d+\s*(?:-\s*\d+)?\s*[- ]?\s*year[s]?\s*\)?", " ", t)
    t = re.sub(r"(?i)\bassignment\b", " ", t)

    # remove ALL digits anywhere
    t = re.sub(r"\d+", "", t)

    # tidy punctuation + whitespace
    t = re.sub(r"[()~]", " ", t)
    t = re.sub(r"\s*-\s*", " ", t)
    t = re.sub(r"\s+", " ", t).strip()

    title_clean = t if t else pd.NA

    return pd.Series(
        {"title_clean": title_clean}
    )
df['title_clean'] = df['filename'].apply(parse_filename)
```

```python
import re

LOCATION_TOKENS = {
    "ab","usa","canada","trinidad","alberta","ontario","bc",
    "calgary","edmonton","vancouver","toronto",
    "zealand","new zealand","nz",
    "medicine hat","medicine","hat",
    "texas","hk","hong kong"
}

# phrases first (e.g., "medicine hat", "new zealand")
_location_terms = sorted(LOCATION_TOKENS, key=len, reverse=True)
LOCATION_RE = re.compile(r"(?i)\b(?:%s)\b" % "|".join(map(re.escape, _location_terms)))

def strip_locations_anywhere(s):
    s = "" if pd.isna(s) else str(s)
    s = LOCATION_RE.sub(" ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

# ✅ remove locations from titles (this fixes rep titles)
df["title_clean"] = df["title_clean"].map(strip_locations_anywhere)
```

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# 1) Domain-specific stopwords (add/remove based on what you saw in cluster keywords)
CUSTOM_STOPWORDS = {
    # boilerplate / branding
    "team","talented","powerful","impact","bright","future","ability","leading","work","working",
    "including","various","ensure","ensures","afraid",

    # super-generic business words that often don't help clustering
    "provide","obtain","information","process","reports","reporting","responsible","support",
    "assist","assistants","office","meeting","events", "going",

    # common scraped artifacts
    "title","location","department","board","committee",

    # overly generic terms
    "jds","docsenior","post","completing",
    "physical","peaks","noise","different","factors",
    "time","deadlines","travel", "intern"
}

# 2) remove known location noise (expand if needed)
LOCATION_TOKENS = {
    "ab","usa","canada","trinidad","alberta","ontario","bc",
    "calgary","edmonton","vancouver","toronto","zealand", "new zealand", "nz", "medicine hat", "medicine","hat",
    "texas", "hk"
}

# 3) remove common boilerplate phrases (exact phrases you saw)
BOILERPLATE_PHRASES = [
    r"\bpowerful impact\b",
    r"\btalented team\b",
    r"\bbright future\b",
    r"\bjoin our team\b",
]

def clean_text(s: str) -> str:
    if pd.isna(s):
        return ""

    s = str(s).lower()

    # remove boilerplate phrases
    for pat in BOILERPLATE_PHRASES:
        s = re.sub(pat, " ", s)

    # remove URLs/emails
    s = re.sub(r"https?://\S+|www\.\S+", " ", s)
    s = re.sub(r"\b[\w\.-]+@[\w\.-]+\.\w+\b", " ", s)

    # normalize whitespace / remove weird characters
    s = s.replace("\n", " ")
    s = re.sub(r"[^a-z\s]", " ", s)      # keep only letters + spaces
    s = re.sub(r"\s+", " ", s).strip()

    # drop short tokens + location tokens
    tokens = [t for t in s.split() if len(t) >= 3 and t not in LOCATION_TOKENS]
    return " ".join(tokens)

# ---- Apply cleaning
TEXT_COL = "text"
df[TEXT_COL] = df[TEXT_COL].fillna("").map(clean_text)

# ---- TF-IDF with combined stopwords
STOPWORDS = list(set(ENGLISH_STOP_WORDS) | CUSTOM_STOPWORDS)
```

```python
import os
import pickle

# Check for cached embeddings to avoid re-computation
embedding_cache_file = 'embeddings_cache.pkl'
X = None

if os.path.exists(embedding_cache_file):
    print(f"Loading cached embeddings from {embedding_cache_file}...")
    with open(embedding_cache_file, 'rb') as f:
        cache_data = pickle.load(f)
        X = cache_data['embeddings']
        print(f"Loaded cached embeddings: {X.shape}")

if X is None:
    print("Computing embeddings...")
    

# --- 2) Google Embeddings (Vertex AI) in batches ---


from vertexai.language_models import TextEmbeddingModel, TextEmbeddingInput
import numpy as np

model = TextEmbeddingModel.from_pretrained("text-embedding-005")

def embed_texts(
    texts,
    task_type="RETRIEVAL_DOCUMENT",   # good general choice for clustering docs
    output_dimensionality=None,
    batch_size=25
):
    vectors = []
    kwargs = dict(output_dimensionality=output_dimensionality) if output_dimensionality else {}

    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]

        inputs = [
            TextEmbeddingInput(task_type=task_type, title="", text=t)
            for t in batch
        ]

        embeddings = model.get_embeddings(inputs, **kwargs)
        vectors.extend([e.values for e in embeddings])

    return np.array(vectors, dtype=np.float32)
    # Save embeddings to cache
    print(f"Saving embeddings to {embedding_cache_file}...")
    with open(embedding_cache_file, 'wb') as f:
        pickle.dump({'embeddings': X}, f)
    print("Embeddings cached for future runs!")

print(f"Embedding matrix shape: {X.shape}")
```

```python
X = embed_texts(df["text"].tolist(), batch_size=25)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# KMeans

k = 25
kmeans = KMeans(n_clusters=k, random_state=42, n_init=30)
df["cluster"] = kmeans.fit_predict(X)

# Evaluate (cosine is better for TF-IDF text)
sil = silhouette_score(X, df["cluster"], metric="cosine")
print("Silhouette (cosine):", sil)

# Cluster sizes
cluster_counts = df["cluster"].value_counts().sort_index()
print(cluster_counts)
print("\nSmall clusters (<10):", (cluster_counts < 10).sum())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    stop_words=STOPWORDS,
    ngram_range=(1,2),
    min_df=2,
    max_df=0.85,
    sublinear_tf=True
)

X_tfidf = vectorizer.fit_transform(df["text"].fillna(""))
terms = np.array(vectorizer.get_feature_names_out())

global_mean = np.asarray(X_tfidf.mean(axis=0)).ravel()

def top_distinctive_terms(cluster_id, top_n=15):
    idx = np.where(df["cluster"].to_numpy() == cluster_id)[0]
    if len(idx) == 0:
        return []
    cluster_mean = np.asarray(X_tfidf[idx].mean(axis=0)).ravel()
    diff = cluster_mean - global_mean
    top_idx = diff.argsort()[::-1][:top_n]
    return terms[top_idx]

for c in sorted(df["cluster"].unique()):
    kws = top_distinctive_terms(c, top_n=15)
    print(f"\nCluster {c} distinctive terms: " + ", ".join(kws))
```

# Labels, Keywords, Job Titles

```python
import re
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.metrics import pairwise_distances

# ----------------------------
# 0) Column names: robust defaults
# ----------------------------
cluster_col = "cluster" if "cluster" in df.columns else "Cluster"
text_col    = "text" if "text" in df.columns else ("full_text" if "full_text" in df.columns else None)
title_col   = "title_clean" if "title_clean" in df.columns else (
    "Unified Job Title" if "Unified Job Title" in df.columns else (
        "Unified Job Title (display)" if "Unified Job Title (display)" in df.columns else None
    )
)

if text_col is None:
    raise ValueError("Couldn't find a text column. Expected df['text'] or df['full_text'].")

if title_col is None:
    raise ValueError("Couldn't find a title column. Expected df['title_clean'], df['Unified Job Title'] or df['Unified Job Title (display)'].")

# ----------------------------
# 1) Display-title cleaning
# ----------------------------
def clean_title_display(x):
    if pd.isna(x):
        return ""
    s = str(x)
    s = s.replace("\t"," ").replace("\n"," ").replace("\r"," ")
    s = re.sub(r"\s+", " ", s).strip()
    s = s.replace("–","-").replace("—","-")   # <-- fixed (was "—" twice)
    s = re.sub(r"\s*-\s*$", "", s).strip()
    s = re.sub(r"\s*(\(|\[).*(\)|\])\s*$", "", s).strip()
    return s

df["Unified Job Title (display)"] = df[title_col].apply(clean_title_display)

# ----------------------------
# 2) TF-IDF fit (global)
# ----------------------------
custom_stop = {
    "working","work","including","ability","required","responsible","responsibilities",
    "experience","knowledge","skills","strong","excellent","must","will","within",
    "ensure","support","provide","team","teams","role","position","etc",
    "job","candidate","employees","employee","business","company","organization",
    "years","year","day","days","new","based"
}
stopwords = ENGLISH_STOP_WORDS.union(custom_stop)

tfidf_vectorizer = TfidfVectorizer(
    stop_words=list(stopwords),
    ngram_range=(1, 2),
    max_df=0.85,
    min_df=3,
)
tfidf_vectorizer.fit(df[text_col].astype(str))

# ----------------------------
# 3) Keyword extraction
# ----------------------------
def extract_cluster_keywords(tfidf, cluster_text, top_n=5):
    response = tfidf.transform([cluster_text])
    feature_names = tfidf.get_feature_names_out()
    if response.nnz == 0:
        return []

    top_idx = np.argsort(response.data)[::-1]
    terms = [feature_names[response.indices[i]] for i in top_idx]

    seen = set()
    uniq_terms = []
    for t in terms:
        if t not in seen:
            uniq_terms.append(t)
            seen.add(t)
        if len(uniq_terms) >= top_n:
            break
    return uniq_terms

# ----------------------------
# 4) Cluster labels (your manual mapping)
# ----------------------------
cluster_label_map = {
    0:  "Applications & Business Systems Analysis",
    1:  "Treasury & Corporate Finance Leadership",
    2:  "Customer Service & Logistics (Supply Chain Execution)",
    3:  "Corporate Admin Support (Office Coordination)",
    4:  "IT Applications Support",
    5:  "Process Engineering",
    6:  "Financial Reporting & Accounting (Controllership)",
    7:  "HR Operations",
    8:  "Plant Operations & HSE / Shift & Lab Support",
    9:  "Corporate Communications & Sustainability/ESG",
    10: "Engineering (Electrical / Instrumentation & Controls)",
    11: "Internal Audit & SOX Compliance",
    12: "Tax & Transfer Pricing",
    13: "Legal (Corporate Counsel / Transactions)",
    14: "Executive & Legal Administrative Support",
    15: "Front Office & Reception",
    16: "Process Safety",
    17: "Maintenance & Reliability Engineering",
    18: "Oracle Finance Systems",
    19: "Enterprise IT / Technology Services",
    20: "Global Supply Chain Planning",
    21: "Procurement / Buying",
    22: "IT Service Desk & End-User Support",
    23: "Mechanical Maintenance Planning",
    24: "Admin Support (Document Control)",
}

# ----------------------------
# 5) Representative titles by centroid (deduped)
# ----------------------------
def get_centroid_representative_titles(
    df, X, centroids, cluster_id: int,
    cluster_col: str, title_col: str,
    n_titles: int = 3, pool_size: int = 30,
    dedupe_casefold: bool = True,
):
    idx = np.where(df[cluster_col].to_numpy() == cluster_id)[0]
    if len(idx) == 0:
        return []

    d = pairwise_distances(
        X[idx], centroids[cluster_id].reshape(1, -1),
        metric="euclidean"
    ).ravel()

    nearest_idx = idx[np.argsort(d)[:min(pool_size, len(idx))]]
    titles = df.loc[nearest_idx, title_col].astype(str).tolist()

    seen = set()
    out = []
    for t in titles:
        t_clean = t.strip()
        if not t_clean:
            continue
        key = t_clean.casefold() if dedupe_casefold else t_clean
        if key in seen:
            continue
        seen.add(key)
        out.append(t_clean)
        if len(out) >= n_titles:
            break
    return out

# ----------------------------
# 6) Build summary table (state-safe)
# ----------------------------
centroids = kmeans.cluster_centers_

# Just in case need manual overrides (exact strings)
# manual_rep_titles = {
#     8: ["PD IT Administrative Assistant", "PD Executive Assistant Global IT", "Human Resources Assistant"],
# }

TITLE_N   = 3
POOL_SIZE = 40
KEYWORD_N = 5

cluster_summaries = []

# Ensure cluster_ids is defined here (and unique)
cluster_ids = sorted(pd.Series(df[cluster_col].dropna().unique()).astype(int).tolist())

for cid in cluster_ids:
    cid_int = int(cid)
    cluster_df = df[df[cluster_col].astype(int) == cid_int].copy()
    label = cluster_label_map.get(cid_int, f"Cluster {cid_int}")

    rep_titles_list = get_centroid_representative_titles(
        df=df,
        X=X,
        centroids=centroids,
        cluster_id=cid_int,
        cluster_col=cluster_col,
        title_col=title_col,
        n_titles=TITLE_N,
        pool_size=POOL_SIZE
    )

    # ✅ apply manual overrides BEFORE joining
    # if cid_int in manual_rep_titles:
    #     rep_titles_list = manual_rep_titles[cid_int][:TITLE_N]

    rep_titles = ", ".join(rep_titles_list)

    combined_text = " ".join(cluster_df[text_col].astype(str).tolist())
    keywords = extract_cluster_keywords(tfidf_vectorizer, combined_text, top_n=KEYWORD_N)
    keyword_str = ", ".join(keywords)

    cluster_summaries.append({
        "Cluster ID": cid_int,
        "Label": label,
        "Count": len(cluster_df),
        f"Key keywords (top {KEYWORD_N})": keyword_str,
        f"Example job titles (top {TITLE_N})": rep_titles
    })

summary_df = (
    pd.DataFrame(cluster_summaries)
      .sort_values("Cluster ID")
      .drop_duplicates(subset=["Cluster ID"], keep="first")  # safety net
      .reset_index(drop=True)
)

pd.set_option("display.max_colwidth", None)
display(summary_df)
```

# Outliers

```python
# Find Distance to Centroids
centers = kmeans.cluster_centers_

# Define FINAL_K based on the number of clusters used in KMeans
FINAL_K = 25 # Or df['cluster'].nunique() if k is not directly available

# Note: Calculating distance for all rows again might be slow if dataset is huge,
# but for 600 rows it's instant. To be efficient, we reuse embeddings.
distances = []
for idx, row in df.iterrows():
    emb = X[idx] # Corrected from embeddings[idx] to X[idx]
    center = centers[int(row['cluster'])] # Use 'cluster' column
    dist = np.linalg.norm(emb - center)
    distances.append(dist)

df['Distance_to_Center'] = distances

# Identify Outliers (Top 5% furthest points in each cluster)
outliers = []
for i in range(FINAL_K):
    cluster_df = df[df['cluster'] == i] # Use 'cluster' column
    if not cluster_df.empty: # Check if cluster_df is not empty
        threshold = cluster_df['Distance_to_Center'].quantile(0.95)
        cluster_outliers = cluster_df[cluster_df['Distance_to_Center'] > threshold]
        outliers.append(cluster_outliers)

outliers_df = pd.concat(outliers)

print("Top Outliers (Jobs that perform unique or mixed functions):")
display(outliers_df[['job_title', 'cluster', 'Distance_to_Center']].sort_values('Distance_to_Center', ascending=False).head(10))
```

```python
import numpy as np
import pandas as pd
from itertools import combinations

# ---- 0) Robust column name ----
cluster_col = "cluster" if "cluster" in df.columns else "Cluster"

# ---- 1) Centroids (from your trained kmeans) ----
centroids = kmeans.cluster_centers_

# ---- 2) Cosine similarity between centroids ----
def cosine_sim_matrix(A: np.ndarray) -> np.ndarray:
    A = np.asarray(A, dtype=float)
    norms = np.linalg.norm(A, axis=1, keepdims=True)
    norms[norms == 0] = 1.0
    A = A / norms
    return A @ A.T

S = cosine_sim_matrix(centroids)  # (k x k)

# ---- 3) Pair table: which clusters are most similar ----
def cluster_similarity_pairs(sim_matrix: np.ndarray, top_n=30, min_sim=0.85):
    k = sim_matrix.shape[0]
    rows = []
    for i, j in combinations(range(k), 2):
        sim = float(sim_matrix[i, j])
        if sim >= min_sim:
            rows.append((i, j, sim))
    pair_df = pd.DataFrame(rows, columns=["Cluster A", "Cluster B", "CosineSimilarity"])
    if len(pair_df) == 0:
        return pair_df
    pair_df = pair_df.sort_values("CosineSimilarity", ascending=False).head(top_n).reset_index(drop=True)
    return pair_df

# ---- 4) Turn similar pairs into "shared job profile groups" (connected components) ----
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0]*n
    def find(self, x):
        while self.parent[x] != x:
            self.parent[x] = self.parent[self.parent[x]]
            x = self.parent[x]
        return x
    def union(self, a, b):
        ra, rb = self.find(a), self.find(b)
        if ra == rb: return
        if self.rank[ra] < self.rank[rb]:
            self.parent[ra] = rb
        elif self.rank[ra] > self.rank[rb]:
            self.parent[rb] = ra
        else:
            self.parent[rb] = ra
            self.rank[ra] += 1

def propose_shared_job_profiles(sim_matrix: np.ndarray, threshold=0.85, min_group_size=2):
    k = sim_matrix.shape[0]
    uf = UnionFind(k)

    for i, j in combinations(range(k), 2):
        if float(sim_matrix[i, j]) >= threshold:
            uf.union(i, j)

    groups = {}
    for c in range(k):
        root = uf.find(c)
        groups.setdefault(root, []).append(c)

    group_list = [sorted(v) for v in groups.values() if len(v) >= min_group_size]
    group_list = sorted(group_list, key=len, reverse=True)
    return group_list

# ---- 5) Pretty output with labels + your summary_df (if exists) ----
def attach_cluster_info(cluster_id: int):
    label = cluster_label_map.get(cluster_id, f"Cluster {cluster_id}") if "cluster_label_map" in globals() else f"Cluster {cluster_id}"

    count = None
    kw = None
    titles = None
    if "summary_df" in globals():
        # try to match your summary_df columns
        row = summary_df[summary_df["Cluster ID"].astype(int) == int(cluster_id)]
        if len(row) == 1:
            row = row.iloc[0]
            count = row.get("Count", None)
            # keyword col name varies
            kw_col = [c for c in summary_df.columns if "Key keywords" in c]
            if kw_col: kw = row.get(kw_col[0], None)
            title_col = [c for c in summary_df.columns if "Example job titles" in c]
            if title_col: titles = row.get(title_col[0], None)

    return label, count, kw, titles

def print_groups(groups):
    out_rows = []
    for gi, g in enumerate(groups, start=1):
        for c in g:
            label, count, kw, titles = attach_cluster_info(c)
            out_rows.append({
                "Group": gi,
                "Cluster": c,
                "Label": label,
                "Count": count,
                "Keywords": kw,
                "Example Titles": titles
            })
    return pd.DataFrame(out_rows)

# ===== RUN IT =====
MIN_SIM = 0.85  # <- adjust: 0.80 more permissive, 0.90 stricter

pair_df = cluster_similarity_pairs(S, top_n=40, min_sim=MIN_SIM)
groups = propose_shared_job_profiles(S, threshold=MIN_SIM, min_group_size=2)
groups_df = print_groups(groups)

print(f"Candidate shared-profile pairs (cosine similarity >= {MIN_SIM}):")
display(pair_df)

print("\nProposed shared job-profile groups (connected clusters above threshold):")
display(groups_df)
```

```python
import numpy as np
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer

# ----------------------------
# A) Differentiating keywords (cluster vs rest)
# ----------------------------
extra_stop = {
    "working","work","including","ability","required","responsible","responsibilities",
    "experience","knowledge","skills","strong","excellent","must","will","within",
    "ensure","support","provide","team","teams","role","position","etc",
    "job","candidate","employees","employee","business","company","organization",
    "years","year","day","days","new","based", "functional"
    "analysis","analyst","manager","senior","lead","director","intern",
    # optional cleanup for your specific output
    "supply","chain"
}

diff_tfidf = TfidfVectorizer(
    stop_words="english",
    max_df=0.8,
    min_df=3,
    ngram_range=(1, 2)
)

X_all = diff_tfidf.fit_transform(df["text"].astype(str))
terms = np.array(diff_tfidf.get_feature_names_out())

keep_mask = np.array([t not in extra_stop for t in terms])
X_all = X_all[:, keep_mask]
terms = terms[keep_mask]

def top_diff_terms_for_cluster(X, terms, mask, top_n=8):
    in_idx = np.where(mask)[0]
    out_idx = np.where(~mask)[0]
    if len(in_idx) == 0 or len(out_idx) == 0:
        return []
    mu_in = np.asarray(X[in_idx].mean(axis=0)).ravel()
    mu_out = np.asarray(X[out_idx].mean(axis=0)).ravel()
    diff = mu_in - mu_out
    top = diff.argsort()[::-1][:top_n]
    return terms[top].tolist()

# ----------------------------
# B) Lexicon (ONE final dict)
# ----------------------------
skill_lexicon_final = {
    "SQL": [r"\bsql\b", r"\bpostgres\b", r"\bmysql\b", r"\bsnowflake\b", r"\bbigquery\b"],
    "Python": [r"\bpython\b", r"\bpandas\b", r"\bnumpy\b"],
    "Excel": [r"\bexcel\b", r"\bpivot table\b", r"\bvlookup\b"],
    "Oracle/ERP": [r"\boracle\b", r"\berp\b", r"\bebs\b", r"\bpeople soft\b", r"\bsap\b"],
    "ITSM/Service Desk": [r"\bitil\b", r"\bincident\b", r"\bservice desk\b", r"\bticket\b", r"\bsla\b"],

    "Engineering/Maintenance": [
    # Industrial maintenance systems / work management language
    r"\b(cmms|work orders?|preventive maintenance|predictive maintenance|reliability|root cause|rca|tpm|loto|lock[- ]out tag[- ]out|plc|scada|hvac|calibration|vibration analysis|thermography)\b",

    # Industrial setting required (avoids “application/system maintenance” in IT postings)
    r"\b(plant|facility|manufactur(ing|e)|production|shop floor)\b(?=.*\b(maintenance|mechanical|electrical|equipment|reliability)\b)",

    # If “maintenance” appears, exclude obvious IT contexts
    r"\bmaintenance\b(?!.*\b(software|application|applications|system|systems|server|servers|network|networks|database|databases|code|platform|website)\b)",
],
"Safety/EHS": [
    r"\b(ehs|hse)\b",
    r"\bsafety\b",
    r"\bloto\b|\block[- ]out tag[- ]out\b",
    r"\bosha\b|\bwhmis\b",
    r"\bhaz(op|ard)\b|\brisk assessment\b",
],

    # tightened Supply Chain
    "Supply Chain": [
        r"\bsupply chain\b",
        r"\b(demand|supply|materials)\s+planning\b",
        r"\bmrp\b|\bmaterials requirements planning\b",
        r"\b(procurement|sourcing|strategic sourcing|purchasing)\b",
        r"\b(warehouse|warehousing|distribution center|dc)\b",
        r"\b(freight|carrier|forwarder|3pl|third[- ]party logistics)\b",
        r"\b(import|export|customs|duty|tariff|broker)\b",
        r"\b(tms|transportation management system)\b",
        r"\binventory\b(?=.*\b(sku|replenish|allocation|fulfillment|warehouse|cycle count|stock)\b)",
        r"\blogistics\b(?=.*\b(freight|warehouse|3pl|customs|import|export|carrier|forwarder|tms|distribution)\b)"
        r"(?!.*\b(travel|itinerary|event|conference|meeting|calendar)\b)",
    ],

    # expanded Audit/SOX
    "Audit/SOX": [
        r"\bsox\b",
        r"\bsarbanes[- ]oxley\b",
        r"\bicfr\b",
        r"\binternal audit\b",
        r"\baudit(ing)? plan\b",
        r"\bwalkthroughs?\b",
        r"\bcontrol testing\b",
        r"\bkey controls?\b",
        r"\bexternal audit\b",
    ],

    "Tax/Transfer Pricing": [r"\btransfer pricing\b", r"\btaxation\b", r"\btax\b"],
    "HR/Payroll": [r"\bpayroll\b", r"\bhr\b", r"\bcompensation\b", r"\bbenefits\b", r"\bemployee relations\b"],
    "ESG/Sustainability": [r"\besg\b", r"\bsustainability\b", r"\bghg\b", r"\bcarbon\b", r"\bcsr\b"],
    "Finance/Reporting": [r"\bfinancial reporting\b", r"\bifrs\b", r"\bgaap\b", r"\bconsolidation\b"],

    # expanded Legal/Contracts
    "Legal/Contracts": [
        r"\blegal counsel\b",
        r"\bcorporate counsel\b",
        r"\bnda\b",
        r"\bmsa\b",
        r"\bsow\b",
        r"\bterms and conditions\b",
        r"\bcontract(s)? negotiation\b",
        r"\bdraft(ing)? (contracts?|agreements?)\b",
        r"\breview(ing)? (contracts?|agreements?)\b",
    ],

    # admin/EA
    "Admin/EA": [
        r"\bexecutive assistant\b|\bea\b",
        r"\badministrative assistant\b|\boffice administrator\b",
        r"\bcalendar management\b|\bscheduling\b",
        r"\btravel arrangements?\b|\bexpense reports?\b",
        r"\bmeeting minutes\b|\bagenda\b",
    ],

    # comms/PR (function-oriented; still may need tightening depending on corpus)
    "Communications/PR": [
    r"\b(corporate|internal|external|global)\s+communications?\b",
    r"\bpublic relations?\b",
    r"\bmedia relations?\b",
    r"\bpress release(s)?\b",
    r"\bcrisis communications?\b",
    r"\bcommunications?\s+(strategy|plan|planning|campaigns?)\b",
    r"\bbrand (messaging|communications?)\b",
    r"\bnewsletter(s)?\b",
],


}

def _hit_rate(text_series: pd.Series, patterns) -> float:
    texts = text_series.astype(str).str.lower()
    pat = "(" + "|".join(patterns) + ")"
    return texts.str.contains(pat, regex=True, na=False).mean()

def cluster_vs_rest_lexicon(df, cluster_col, text_col, lexicon, cluster_id):
    in_mask = df[cluster_col].to_numpy() == cluster_id
    out_mask = ~in_mask
    out = []
    for cat, pats in lexicon.items():
        r_in = _hit_rate(df.loc[in_mask, text_col], pats)
        r_out = _hit_rate(df.loc[out_mask, text_col], pats)
        diff = r_in - r_out
        lift = (r_in + 1e-6) / (r_out + 1e-6)
        out.append((cat, r_in, r_out, diff, lift))
    return pd.DataFrame(out, columns=["Category","Hit_in","Hit_out","Diff(in-out)","Lift(in/out)"])

# ----------------------------
# C) Title-based seniority + role family
# ----------------------------
seniority_priority = [
    ("Director+", r"\bdirector\b|\bvp\b|\bvice president\b|\bhead\b|\bchief\b"),
    ("Manager",   r"\bmanager\b|\bmgr\b|\bsupervisor\b"),
    ("Lead/Principal", r"\blead\b|\bprincipal\b|\bstaff\b|\bteam lead\b|\btech lead\b"),
    ("Senior",    r"\bsenior\b|\bsr\\.?\b"),
    ("Junior/Associate", r"\bjunior\b|\bassociate\b|\bentry\b|\bgraduate\b|\bnew grad\b"),
    ("Intern",    r"\bintern\b|\bco-?op\b"),
]

def assign_seniority(title: str) -> str:
    t = str(title).lower()
    for name, pat in seniority_priority:
        if re.search(pat, t):
            return name
    return "Unspecified"

def seniority_dist_exclusive(title_series: pd.Series):
    cats = title_series.apply(assign_seniority)
    return cats.value_counts(normalize=True).to_dict()

role_priority = [
    ("Legal/Counsel", r"\bcounsel\b|\battorney\b|\blegal\b"),
    ("Audit", r"\bauditor\b|\baudit\b"),
    ("Tax", r"\btax\b|\btransfer pricing\b"),
    ("HR/People", r"\bhuman resources\b|\bhr\b|\bpayroll\b|\btalent\b"),
    ("Sustainability/ESG", r"\besg\b|\bsustainability\b|\bcsr\b"),
    ("Supply Chain/Logistics", r"\bsupply chain\b|\blogistics\b|\bplanner\b|\bprocurement\b|\binventory\b"),
    ("IT/Systems", r"\bit\b|\bsystems?\b|\bapplication(s)?\b|\bservice desk\b|\boracle\b|\berp\b|\binfrastructure\b"),
    ("Finance/Accounting", r"\bfinance\b|\baccounting\b|\bcontroller\b|\breporting\b"),
    ("Engineering/Maintenance", r"\bengineer\b|\bmechanical\b|\belectrical\b|\bmaintenance\b|\breliability\b|\bplant\b"),
    ("Admin/EA", r"\badministrative\b|\badmin\b|\bexecutive assistant\b|\boffice coordinator\b|\breceptionist\b"),
    ("Analyst", r"\banalyst\b"),
    ("Coordinator/Specialist", r"\bcoordinator\b|\bspecialist\b"),
]

def assign_role_family(title: str) -> str:
    t = str(title).lower()
    for name, pat in role_priority:
        if re.search(pat, t):
            return name
    return "Other/Unspecified"

def role_family_dist_exclusive(title_series: pd.Series):
    cats = title_series.apply(assign_role_family)
    return cats.value_counts(normalize=True).to_dict()

# ----------------------------
# D) Build comparison table per cluster
# ----------------------------
rows = []
clusters = sorted(df["cluster"].dropna().unique().astype(int))

for i in clusters:
    mask = (df["cluster"].to_numpy() == i)

    diff_terms = top_diff_terms_for_cluster(X_all, terms, mask, top_n=8)

    lex_df = cluster_vs_rest_lexicon(df, "cluster", "text", skill_lexicon_final, i)
    lex_df = lex_df[
        (lex_df["Diff(in-out)"] >= 0.05) &
        (lex_df["Lift(in/out)"] >= 1.20) &
        (lex_df["Hit_in"] >= 0.10)
    ].sort_values(["Diff(in-out)", "Lift(in/out)"], ascending=False).head(5)

    top_skills_str = (
        ", ".join([f"{r.Category} ({r.Hit_in:.0%}, lift {r['Lift(in/out)']:.2f})" for _, r in lex_df.iterrows()])
        if len(lex_df) else
        "N/A (no distinctive skill signals vs rest)"
    )

    titles_i = df.loc[mask, "Unified Job Title (display)"]

    senior_dist = seniority_dist_exclusive(titles_i)
    top_senior = sorted(senior_dist.items(), key=lambda x: x[1], reverse=True)[:3]

    role_dist = role_family_dist_exclusive(titles_i)
    top_roles = sorted(role_dist.items(), key=lambda x: x[1], reverse=True)[:3]

    rows.append({
        "Cluster ID": i,
        "Differentiating keywords (vs rest)": ", ".join(diff_terms),
        "Top skill/resp signals (lift-filtered)": top_skills_str,
        # "Typical seniority (title-based)": ", ".join([f"{k} ({v:.0%})" for k, v in top_senior]),
        "Typical role family (title-based)": ", ".join([f"{k} ({v:.0%})" for k, v in top_roles]),
        "Count": int(mask.sum())
    })

cluster_diff_df = pd.DataFrame(rows).sort_values("Cluster ID")
pd.set_option("display.max_colwidth", None)
display(cluster_diff_df)
```

```python

# ----------------------------
# 2) Extract Individual Skills for Each Employee
# ----------------------------

# Use the existing skill lexicon from the cluster analysis
skill_lexicon_individual = {
    "SQL": [r"\bsql\b", r"\bpostgres\b", r"\bmysql\b", r"\bsnowflake\b", r"\bbigquery\b"],
    "Python": [r"\bpython\b", r"\bpandas\b", r"\bnumpy\b"],
    "Excel": [r"\bexcel\b", r"\bpivot table\b", r"\bvlookup\b"],
    "Oracle/ERP": [r"\boracle\b", r"\berp\b", r"\bebs\b", r"\bpeople soft\b", r"\bsap\b"],
    "ITSM/Service Desk": [r"\bitil\b", r"\bincident\b", r"\bservice desk\b", r"\bticket\b", r"\bsla\b"],
    "Engineering/Maintenance": [
        r"\b(cmms|work orders?|preventive maintenance|predictive maintenance|reliability|root cause|rca|tpm|loto|lock[- ]out tag[- ]out|plc|scada|hvac|calibration|vibration analysis|thermography)\b",
        r"\b(plant|facility|manufactur(ing|e)|production|shop floor)\b(?=.*\b(maintenance|mechanical|electrical|equipment|reliability)\b)",
        r"\bmaintenance\b(?!.*\b(software|application|applications|system|systems|server|servers|network|networks|database|databases|code|platform|website)\b)",
    ],
    "Safety/EHS": [
        r"\b(ehs|hse)\b", r"\bsafety\b", r"\bloto\b|\block[- ]out tag[- ]out\b",
        r"\bosha\b|\bwhmis\b", r"\bhaz(op|ard)\b|\brisk assessment\b",
    ],
    "Supply Chain": [
        r"\bsupply chain\b", r"\b(demand|supply|materials)\s+planning\b",
        r"\bmrp\b|\bmaterials requirements planning\b", r"\b(procurement|sourcing|strategic sourcing|purchasing)\b",
        r"\b(warehouse|warehousing|distribution center|dc)\b", r"\b(freight|carrier|forwarder|3pl|third[- ]party logistics)\b",
        r"\b(import|export|customs|duty|tariff|broker)\b", r"\b(tms|transportation management system)\b",
    ],
    "Audit/SOX": [
        r"\bsox\b", r"\bsarbanes[- ]oxley\b", r"\bicfr\b", r"\binternal audit\b",
        r"\baudit(ing)? plan\b", r"\bwalkthroughs?\b", r"\bcontrol testing\b", r"\bkey controls?\b",
    ],
    "Tax/Transfer Pricing": [r"\btransfer pricing\b", r"\btaxation\b", r"\btax\b"],
    "HR/Payroll": [r"\bpayroll\b", r"\bhr\b", r"\bcompensation\b", r"\bbenefits\b", r"\bemployee relations\b"],
    "ESG/Sustainability": [r"\besg\b", r"\bsustainability\b", r"\bghg\b", r"\bcarbon\b", r"\bcsr\b"],
    "Finance/Reporting": [r"\bfinancial reporting\b", r"\bifrs\b", r"\bgaap\b", r"\bconsolidation\b"],
    "Legal/Contracts": [
        r"\blegal counsel\b", r"\bcorporate counsel\b", r"\bnda\b", r"\bmsa\b", r"\bsow\b",
        r"\bterms and conditions\b", r"\bcontract(s)? negotiation\b", r"\bdraft(ing)? (contracts?|agreements?)\b",
    ],
    "Admin/EA": [
        r"\bexecutive assistant\b|\bea\b", r"\badministrative assistant\b|\boffice administrator\b",
        r"\bcalendar management\b|\bscheduling\b", r"\btravel arrangements?\b|\bexpense reports?\b",
    ],
    "Communications/PR": [
        r"\b(corporate|internal|external|global)\s+communications?\b", r"\bpublic relations?\b",
        r"\bmedia relations?\b", r"\bpress release(s)?\b", r"\bcrisis communications?\b",
    ],
    "Data Analysis": [r"\bdata analysis\b", r"\bdata analytics\b", r"\bdata modeling\b", r"\bdata visualization\b"],
    "Project Management": [r"\bproject management\b", r"\bproject manager\b", r"\bpmo\b", r"\bstakeholder management\b"],
    "IT/Business Systems": [r"\bit systems\b", r"\bbusiness systems\b", r"\bsystems analyst\b", r"\bapplication analyst\b"],
}

def extract_skills_for_employee(text, lexicon):
    """Extract skills for a single employee/job posting"""
    if pd.isna(text):
        return []
    text_lower = str(text).lower()
    skills_found = []
    for skill_name, patterns in lexicon.items():
        for pattern in patterns:
            if re.search(pattern, text_lower, re.IGNORECASE):
                skills_found.append(skill_name)
                break  # Avoid duplicate skills
    return skills_found

# Extract skills for each employee
print("Extracting individual skills for each employee...")
df['Individual_Skills'] = df['text'].apply(lambda x: extract_skills_for_employee(x, skill_lexicon_individual))
df['Skills_Count'] = df['Individual_Skills'].apply(len)
df['Skills_String'] = df['Individual_Skills'].apply(lambda x: ", ".join(x) if x else "None")

print(f"\nSkills extraction complete!")
print(f"Average skills per employee: {df['Skills_Count'].mean():.2f}")
print(f"Max skills for one employee: {df['Skills_Count'].max()}")
print(f"Employees with no skills identified: {(df['Skills_Count'] == 0).sum()}")
```

```python

# ----------------------------
# 4) Find Top 3 Most Similar Jobs for Each Employee
# ----------------------------

print("Calculating similarity matrix for all employees...")

# Calculate pairwise cosine similarity using the embeddings X
def cosine_similarity_matrix(X):
    """Calculate cosine similarity matrix"""
    X_norm = X / np.linalg.norm(X, axis=1, keepdims=True)
    return X_norm @ X_norm.T

sim_matrix = cosine_similarity_matrix(X)

# For each employee, find top 3 most similar (excluding self)
def get_top_similar(emp_idx, sim_matrix, n=3):
    """Get indices of top n most similar employees"""
    sim_scores = sim_matrix[emp_idx].copy()
    sim_scores[emp_idx] = -1  # Exclude self
    top_indices = np.argsort(sim_scores)[::-1][:n]
    return top_indices, sim_scores[top_indices]

# Create columns for similar jobs
df['Similar_Employee_1'] = ''
df['Similar_Employee_1_Score'] = 0.0
df['Similar_Employee_2'] = ''
df['Similar_Employee_2_Score'] = 0.0
df['Similar_Employee_3'] = ''
df['Similar_Employee_3_Score'] = 0.0

print("Finding top 3 similar jobs for each employee...")

for i in range(len(df)):
    top_idx, scores = get_top_similar(i, sim_matrix, n=3)
    
    df.at[i, 'Similar_Employee_1'] = df.iloc[top_idx[0]]['Employee_ID']
    df.at[i, 'Similar_Employee_1_Score'] = round(scores[0], 4)
    
    df.at[i, 'Similar_Employee_2'] = df.iloc[top_idx[1]]['Employee_ID']
    df.at[i, 'Similar_Employee_2_Score'] = round(scores[1], 4)
    
    df.at[i, 'Similar_Employee_3'] = df.iloc[top_idx[2]]['Employee_ID']
    df.at[i, 'Similar_Employee_3_Score'] = round(scores[2], 4)

print("Similarity calculation complete!")
```

```python

# ----------------------------
# 5) Display Employee with Similar Jobs
# ----------------------------

print("=" * 120)
print("EMPLOYEES WITH THEIR INDIVIDUAL SKILLS AND TOP 3 MOST SIMILAR JOBS")
print("=" * 120)

def display_employee_with_similar(emp_id):
    """Display an employee and their similar jobs"""
    emp = df[df['Employee_ID'] == emp_id].iloc[0]
    
    print(f"\n{'='*120}")
    print(f"🆔 EMPLOYEE: {emp['Employee_ID']}")
    print(f"   Job Title: {emp['title_clean']}")
    print(f"   Cluster: {emp['cluster']} - {cluster_label_map.get(emp['cluster'], 'Unknown')}")
    print(f"   Skills ({emp['Skills_Count']}): {emp['Skills_String']}")
    print(f"\n   📊 TOP 3 MOST SIMILAR JOBS:")
    
    for i, (col_id, col_score) in enumerate([
        ('Similar_Employee_1', 'Similar_Employee_1_Score'),
        ('Similar_Employee_2', 'Similar_Employee_2'),
        ('Similar_Employee_3', 'Similar_Employee_3_Score')
    ], 1):
        similar_emp_id = emp[col_id]
        similar_emp = df[df['Employee_ID'] == similar_emp_id].iloc[0]
        score = emp[f'Similar_Employee_{i}_Score']
        
        print(f"\n   {i}. {similar_emp_id} (Similarity: {score:.4f})")
        print(f"      Title: {similar_emp['title_clean']}")
        print(f"      Cluster: {similar_emp['cluster']} - {cluster_label_map.get(similar_emp['cluster'], 'Unknown')}")
        print(f"      Skills: {similar_emp['Skills_String']}")

# Display first 5 employees with their similar jobs
for emp_id in df['Employee_ID'].head(5).values:
    display_employee_with_similar(emp_id)

print(f"\n{'='*120}")
print("SUMMARY STATISTICS")
print(f"{'='*120}")
print(f"Total Employees: {len(df)}")
print(f"Total Unique Skills Categories: {len(skill_lexicon_individual)}")
print(f"Average Skills per Employee: {df['Skills_Count'].mean():.2f}")
print(f"Average Similarity Score (Top Match): {df['Similar_Employee_1_Score'].mean():.4f}")
```

```python
# -------------------------------------------------------
# Generate 2D Coordinates for Visualization
# -------------------------------------------------------
import numpy as np

try:
    import umap
    reducer = umap.UMAP(n_components=2, random_state=42, n_neighbors=15, min_dist=0.1, metric='cosine')
    coords_2d = reducer.fit_transform(X)
    print('UMAP coordinates generated successfully')
except ImportError:
    from sklearn.decomposition import PCA
    reducer = PCA(n_components=2, random_state=42)
    coords_2d = reducer.fit_transform(X)
    print('PCA coordinates generated (UMAP not available)')

# Add to dataframe
df['x'] = coords_2d[:, 0]
df['y'] = coords_2d[:, 1]

# Normalize to [-50, 50] range for visualization
for col in ['x', 'y']:
    df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
    df[col] = df[col] * 100 - 50

print(f'Coordinates range: x=[{df["x"].min():.2f}, {df["x"].max():.2f}], y=[{df["y"].min():.2f}, {df["y"].max():.2f}]')
```
